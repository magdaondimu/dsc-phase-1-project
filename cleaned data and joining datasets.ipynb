CHANGE THE COLUMN NAMES FROM ONE DATASET TO JOIN WITH ANOTHER DATASET WITH THE SAME NAMES.

```python
tn_movie.rename(
    columns={'release_date': 'year', 'movie': 'title', 'worldwide_gross': 'foreign_gross'}, 
    inplace=True,
)
tmdb_movies = tmdb_movies.rename(columns={'release_date': 'year'})
title_basics = title_basics.rename(columns={'start_year': 'year', 'primary_title': 'title'})
```

CONCANTENETE THE TWO DATASETS.

```python
joined_movies = pd.concat([bom_movie, tn_movie, tmdb_movies, title_basics], ignore_index=True)
joined_movies
```

DROP THE COLUMNS NOT NEEDED IN THE NEW DATASET

```python
joined_movies.drop(columns=['id', 'production_budget', 'genre_ids', 'original_language', 'original_title','vote_average', 
                            'vote_count','tconst','runtime_minutes'], inplace=True)
joined_movies
```
